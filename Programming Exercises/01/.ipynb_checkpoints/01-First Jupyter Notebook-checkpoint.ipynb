```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Generating random data in a normal distribution

```python
np.random.seed(239)
x = np.linspace(0, 50, 100)
y = 4 * x + 2 + np.random.normal(0, 1, 100)
```

Creating a scatter-plot graph

```python
plt.plot(x, 4 * x + 2, label='True Relationship', color='blue', linestyle='--')

plt.title('Simulated Data Scatter and Line Plot')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.legend()
plt.grid(True)
plt.show()
```
